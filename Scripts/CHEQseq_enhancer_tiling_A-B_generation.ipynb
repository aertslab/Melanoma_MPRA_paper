## Score full amplified regions with Cluster-Buster

```bash
# Load Cluster-Buster module.
module load Cluster-Buster/20180702-ivybridge-intel-2018a


score_full_amplified_regions_with_cluster_buster () {
    local cluster_buster_motif_filename="${1}"
    local cbust_BED_output_file="${2}"
    local cluster_threshold="${3}"
    local motif_threshold="${4}"
    
    if [ "${#@}" -ne 4 ] ; then
        printf 'Usage:  score_full_amplified_regions_with_cluster_buster \\\n'
        printf '            cluster_buster_motif_filename \\\n'
        printf '            cbust_BED_output_file \\\n'
        printf '            cluster_threshold \\\n'
        printf '            motif_threshold\n'
        return 1;
    fi

    printf 'Score amplified regions with "%s": c=%s m=%s.\n' \
        "${motif_name}" \
        "${cluster_threshold}" \
        "${motif_threshold}"
    
    cbust \
        -c ${cluster_threshold} \
        -m ${motif_threshold}  \
        -r 1000000 \
        -G 1 \
        -f 5 \
        "${cluster_buster_motif_filename}" \
        "${amplified_cbust_seq_names_fasta_filename}" \
      > "${cbust_BED_output_file}"

    printf 'Output Cluster-Buster BED file: %s' \
        "${cbust_BED_output_file}"
}
```

```bash
import hashlib
import numpy as np
import pandas as pd
import re
import subprocess

from collections import OrderedDict

from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from Bio.Alphabet import generic_dna


class Mutation:
    @staticmethod
    def from_str(mutation_str):
        chrom, genomic_position_and_ref_to_mut = mutation_str.split(':')
        genomic_position, ref_to_mut = genomic_position_and_ref_to_mut.split('__')
        genomic_position = int(genomic_position)
        ref_nucleotide, mut_nucleotide = ref_to_mut.split('_')

        return Mutation(chrom, genomic_position, ref_nucleotide, mut_nucleotide)
        
    def __init__(self, chrom, genomic_position, ref_nucleotide, mut_nucleotide):
        self.chrom = chrom
        # One-based genomic position.
        self.genomic_position = genomic_position
        self.ref_nucleotide = ref_nucleotide
        self.mut_nucleotide = mut_nucleotide
    
    def __str__(self):
        # Add introduced mutation to the sequence description:
        #   chromosome : genomic position __ reference nucleotide _ mutated nucleotide
        return '{0:s}:{1:d}__{2:s}_{3:s}'.format(
            self.chrom,
            self.genomic_position,
            self.ref_nucleotide,
            self.mut_nucleotide
        )
    
    def __eq__(self, other):
        if self.chrom != other.chrom:
            raise(
                ValueError('Mutation "{0:s}" and "{1:s}" have a different chromosome.'.format(
                    str(self),
                    str(other)
                ))
            )
        
        return self.genomic_position == other.genomic_position
    
    def __lt__(self, other):
        if self.chrom != other.chrom:
            raise(
                ValueError('Mutation "{0:s}" and "{1:s}" have a different chromosome.'.format(
                    str(self),
                    str(other)
                ))
            )
        
        return self.genomic_position < other.genomic_position
    
    def in_interval(self, genommic_interval_start, genomic_interval_end):
        # All coordinates are one-based.
        return (
            self.genomic_position >= genommic_interval_start
            and self.genomic_position <= genomic_interval_end
        )


class SequenceNameAndDescription:
    @staticmethod
    def from_str(seq_name, seq_description):
        seq_name_regex = '(?P<chrom>[^:]+):(?P<genomic_start_position>[0-9]+)-(?P<genomic_end_position>[0-9]+)(@@(?P<gene>[^=]+))?(==(?P<wt_or_mut>(wt|mut)))?'
        
        seq_name_regex_results = re.match(seq_name_regex, seq_name)
        
        if seq_name_regex_results is None:
            raise ValueError('Sequence name "{0:s}" is not in the following format: "chr:start-end", "chr:start-end@@gene", "chr:start-end@@gene==wt" or "chr:start-end@@gene==mut"'.format(seq_name))
        
        chrom = seq_name_regex_results.group('chrom')
        genomic_start_position = int(seq_name_regex_results.group('genomic_start_position'))
        genomic_end_position = int(seq_name_regex_results.group('genomic_end_position'))
        gene = seq_name_regex_results.group('gene')
        wt_or_mut = seq_name_regex_results.group('wt_or_mut')
        
        motif_name__mutations = seq_description.split(' ')
        
        if motif_name__mutations[0] == seq_name:
            # Remove sequence name from description, it it is there.
            del motif_name__mutations[0]
        
        if len(motif_name__mutations) >= 1:
            # Extract motif name from description.
            motif_name = motif_name__mutations[0]
        else:
            motif_name = None

        if len(motif_name__mutations) >= 2:
            # Extract already introduced mutations from sequence description.
            mutations = [Mutation.from_str(mutation_str) for mutation_str in motif_name__mutations[1:]]
        else:
            mutations = []
        
        return SequenceNameAndDescription(
            chrom,
            genomic_start_position,
            genomic_end_position,
            gene,
            motif_name,
            mutations
        )
    
    def __init__(self, chrom, genomic_start_position, genomic_end_position, gene, motif_name, mutations):
        self.chrom = chrom
        # One-base genomic start coordinate of the sequence.
        self.genomic_start_position = genomic_start_position
        # One-base genomic end coordinate of the sequence.
        self.genomic_end_position = genomic_end_position
        self.gene = gene
        self.motif_name = motif_name
        
        # Filter out mutations that do not fit in the genomic interval of the sequence
        # and sort them.
        
        self.mutations = mutations
        self.filter_mutations()
    
    def filter_mutations(self):
        # Filter out mutations that do not fit in the genomic interval of the sequence
        # and sort them.
        self.mutations = list(sorted(
            [mutation
             for mutation in self.mutations
             if mutation.in_interval(genommic_interval_start=self.genomic_start_position,
                                     genomic_interval_end=self.genomic_end_position)
            ]))
    
    def add_mutation_to_sequence_description(self, mutation):
        if mutation.in_interval(genommic_interval_start=self.genomic_start_position,
                                genomic_interval_end=self.genomic_end_position):
            # Add mutation if it fits in the genomic interval.
            self.mutations.append(mutation)
        
            # Sort the mutations.
            #self.mutations = list(sorted(self.mutations))
            self.filter_mutations()
    
    def has_mutations(self):
        return True if self.mutations else False
        
    def get_sequence_name_and_description(self, with_wt_or_mut):
        if with_wt_or_mut:
            sequence_name = '{0:s}:{1:d}-{2:d}{3:s}=={4:s}'.format(
                self.chrom,
                self.genomic_start_position,
                self.genomic_end_position,
                '@@' + self.gene if self.gene else '',
                'mut' if len(self.mutations) >= 1 else 'wt'
            )
        else:
            sequence_name = '{0:s}:{1:d}-{2:d}{3:s}'.format(
                self.chrom,
                self.genomic_start_position,
                self.genomic_end_position,
                '@@' + self.gene if self.gene else ''
            )
        
        sequence_description = ''
        
        if self.motif_name:
            sequence_description = self.motif_name

        if self.mutations:
            sequence_description = '{0:s} {1:s}'.format(
                sequence_description,
                ' '.join([str(mutation)
                          for mutation in self.mutations])
            )
        
        # If with_wt_or_mut is True and the sequence is wildtype (no mutations),
        # keep the sequence description empty (no motif name).
        if with_wt_or_mut and len(self.mutations) == 0:
            sequence_description = ''
        
        return (sequence_name, sequence_description)
        

class MutateFastaSequences:
    @staticmethod
    def load_fasta_from_file(fasta_input_filename, motif_name):
        mut_fasta_dict = OrderedDict()

        with open(fasta_input_filename, 'r') as fasta_input_fh:
            # Read all FASTA sequences in a dictionary with:
            #   - the sequence name as key
            #   - a SeqRecord as value which contains:
            #       - a mutable sequence
            #       - description: motif name plus introcuded mutations.
            for record in SeqIO.parse(fasta_input_fh, 'fasta', generic_dna):
                record.seq = record.seq.tomutable()
                record.description = motif_name
                mut_fasta_dict[record.id] = record
        
        return MutateFastaSequences(mut_fasta_dict=mut_fasta_dict)
    
    def __init__(self, mut_fasta_dict):
        self.mut_fasta_dict = mut_fasta_dict
    
    def mutate_fasta(self, seq_name, mut_genomic_position, mut_relative_position, ref_nucleotide, mut_nucleotide):
        # Coordinates are zero-based.
        
        # Get the original (could be the reference, unless already mutated before)
        # nucleotide before mutating it.
        original_nucleotide = self.mut_fasta_dict[seq_name].seq[mut_relative_position]

        # Make sure the reference and mutated nucleotide has the same case
        # (lower or upper) than the original nucleotide in the FASTA file.
        ref_nucleotide, mut_nucleotide = ((ref_nucleotide.lower(), mut_nucleotide.lower())
                                          if original_nucleotide == original_nucleotide.lower()
                                          else (ref_nucleotide.upper(), mut_nucleotide.upper()))
        
        # Introduce the mutation at the specified position.
        self.mut_fasta_dict[seq_name].seq[mut_relative_position] = mut_nucleotide
        
        # Create mutation object.
        current_mut = Mutation(
            chrom=seq_name.split(':')[0],
            genomic_position=mut_genomic_position + 1,
            ref_nucleotide=ref_nucleotide,
            mut_nucleotide=mut_nucleotide
        )

        # Parse sequence name and sequence description.
        sequence_name_and_description = SequenceNameAndDescription.from_str(
            seq_name=seq_name,
            seq_description=self.mut_fasta_dict[seq_name].description
        )
        
        # Add the new mutation.
        sequence_name_and_description.add_mutation_to_sequence_description(current_mut)
        
        # Get the new sequence description.
        self.mut_fasta_dict[seq_name].description = sequence_name_and_description.get_sequence_name_and_description(
            with_wt_or_mut=False
        )[1]
    
    def has_sequence(self, seq_name):
        return seq_name in self.mut_fasta_dict
    
    def get_sequence(self, seq_name):
        return self.mut_fasta_dict.get(seq_name)

    def write_fasta_to_file(self, fasta_output_filename):
        # Write all (mutated) FASTA sequences to a file.
        with open(fasta_output_filename, 'w') as fasta_output_fh:
            for seq_name, record in self.mut_fasta_dict.items():
                SeqIO.write(record, fasta_output_fh, "fasta")

                
class Motif:
    @staticmethod
    def load_motif_from_file(motif_filename):
        with open(motif_filename, 'r') as motif_fh:
            motif_name = motif_fh.readline().lstrip('>').rstrip('\n')
            
            motif_fh.seek(0)
            
            # Read Cluster-Buster motif file in numpy array:
            #   - skip ">motif_name" line
            #   - Read numbers as floats
            motif = np.loadtxt(fname=motif_fh,
                                  dtype=np.float,
                                  delimiter='\t',
                                  skiprows=1)
            
            return Motif(motif, motif_name)
        
        return None
    
    def __init__(self, motif, motif_name):
        self.motif = motif
        self.motif_name = motif_name
        
        # Nucleotide order.
        nucleotide_order = np.array(['A', 'C', 'G', 'T'])
        
        # Normalize motif by dividing each value on a row by the row sum.
        normalized_motif = self.motif / self.motif.sum(axis=1).reshape(-1, 1)
        
        # Get index (nucleotide) of maximum value for each position in the motif.
        max_idx_per_position = normalized_motif.argmax(axis=1).reshape(-1, 1)

        # Get index (nucleotide) of minimum value in each position in the motif.
        min_idx_per_position = normalized_motif.argmin(axis=1).reshape(-1, 1)

        # Maximum value for each position in the motif.
        max_per_position = normalized_motif[
            (np.arange(0, normalized_motif.shape[0]).reshape(-1, 1),
             max_idx_per_position
            )    
        ]

        # Minimum value for each position in the motif.
        min_per_position = normalized_motif[
            (np.arange(0, normalized_motif.shape[0]).reshape(-1, 1),
             min_idx_per_position
            )    
        ]
        
        # Positions in motif from high to low sorted on the maximum value per row.
        self.max_idx_per_position_high_to_low = max_per_position.argsort(axis=0)[::-1]

        # Most likely nucleotide for each of those positions (mutate motif position from this nucleotide).
        self.most_likely_nucleotide_per_position_high_to_low = nucleotide_order[
            max_idx_per_position[self.max_idx_per_position_high_to_low].reshape(-1, 1)
        ]

        # Least likely nucleotide for each of those positions (mutate motif position to this nucleotide).
        self.least_likely_nucleotide_per_position_high_to_low = nucleotide_order[
            min_idx_per_position[self.max_idx_per_position_high_to_low].reshape(-1, 1)
        ]
    
    def destroy_motif_in_fasta(self, mut_fasta_sequences, seq_name, genomic_start,
                               relative_start, strand, motif_sequence, max_nbr_mutations=2):
        if self.motif.shape[0] != len(motif_sequence):
            raise ValueError(
                'Sequence length ({0:d}) should match length of the motif ({1:d}).'.format(
                    len(sequence),
                    self.motif.shape[0]
                )
            )

        if not mut_fasta_sequences.has_sequence(seq_name):
            raise ValueError('Sequence name ({0:d}) not found in FASTA sequences.'.format(
                    seq_name
                )
            )

        # Keep track of the number of introduced mutations.
        current_nbr_mutations = 0
        
        # Make motif sequence mutatable.
        mut_motif_sequence = Seq(motif_sequence, generic_dna).tomutable()

        # Try to mutate strongest motif positions first.
        for position_in_motif, most_likely_nucleotide, least_likely_nucleotide in zip(
                self.max_idx_per_position_high_to_low,
                self.most_likely_nucleotide_per_position_high_to_low,
                self.least_likely_nucleotide_per_position_high_to_low):
            # Remove encapsulating numpy arrays.
            position_in_motif = np.asscalar(position_in_motif)
            most_likely_nucleotide = np.asscalar(most_likely_nucleotide)
            least_likely_nucleotide = np.asscalar(least_likely_nucleotide)

            # Only introduce max_nbr_mutations of mutations.
            if current_nbr_mutations < max_nbr_mutations:
                # Mutate sequence.
                if mut_motif_sequence[position_in_motif].upper() == most_likely_nucleotide:                   
                    if strand == '-':
                        mut_genomic_position = genomic_start + (len(motif_sequence) - 1) - position_in_motif
                        mut_relative_position = relative_start + (len(motif_sequence) - 1) - position_in_motif
                        # Take reverse complement for reference and mutation.
                        ref_nucleotide = str(Seq(mut_motif_sequence[position_in_motif], generic_dna).reverse_complement())
                        mut_nucleotide = str(Seq(least_likely_nucleotide, generic_dna).reverse_complement())
                    else:
                        mut_genomic_position = genomic_start + position_in_motif
                        mut_relative_position = relative_start + position_in_motif
                        ref_nucleotide = mut_motif_sequence[position_in_motif]
                        mut_nucleotide = least_likely_nucleotide
                    
                    # Sequence of motif with the intruduced mutation.
                    mut_motif_sequence[position_in_motif] = least_likely_nucleotide
                    
                    mut_fasta_sequences.mutate_fasta(seq_name,
                                                     mut_genomic_position,
                                                     mut_relative_position,
                                                     ref_nucleotide,
                                                     mut_nucleotide)
                    
                    current_nbr_mutations += 1
            else:
                print(motif_sequence, mut_motif_sequence)
                break

                
class CbustBED:
    @staticmethod
    def load_motifs_only_from_cbust_bed_from_file(cbust_bed_filename):
        with open(cbust_bed_filename, 'r') as cbust_bed_fh:
            cbust_bed_df = pd.read_csv(
                filepath_or_buffer=cbust_bed_fh,
                sep='\t',
                header=0,
                engine='c'
            )
            
            # Fix chromosome header column name.
            cbust_bed_df = cbust_bed_df.rename(columns={'# chrom': 'chrom'})
            
            cbust_motif_only_df = cbust_bed_df[cbust_bed_df.cluster_or_motif == 'motif'][[
                'chrom',
                'genomic_start__bed',
                'genomic_end__bed',
                'strand',
                'seq_name',
                'relative_start__bed',
                'relative_end__bed',
                'motif_sequence'
            ]]
            
            return CbustBED(cbust_motif_only_df=cbust_motif_only_df)
    
    def __init__(self, cbust_motif_only_df):
        self.cbust_motif_only_df = cbust_motif_only_df
    

class Tile():
    @staticmethod
    def make_tiles_from_sequence(sequence_record, first_tile_start_position, tile_size,
                                 tile_step_size, minimum_tile_size, prefix_seq, suffix_seq):
        sequence_length = len(sequence_record.seq)
        
        for tile_start in range(first_tile_start_position, sequence_length, tile_step_size):
            tile_end = (tile_start + tile_size
                        if tile_start + tile_size < sequence_length
                        else sequence_length)

            # Only generate tiles that are at least minimum_tile_size.
            if tile_end - tile_start >= minimum_tile_size:
                print(tile_start)
                yield Tile(sequence_record=sequence_record,
                           tile_start=tile_start,
                           tile_end=tile_end,
                           prefix_seq=prefix_seq,
                           suffix_seq=suffix_seq
                          )
    @staticmethod
    def make_tile_from_sequence_with_adapters(sequence_record_with_adapters, prefix_seq, suffix_seq):
        # Get parse info from sequence name and sequence description of tiled sequence record with adapters.
        sequence_name_and_description = SequenceNameAndDescription.from_str(
            seq_name=sequence_record_with_adapters.name,
            seq_description=sequence_record_with_adapters.description
        )

        # Get new sequence name and description based on the new genomic coordinates and filtering of mutations.
        seq_name, seq_description = sequence_name_and_description.get_sequence_name_and_description(
            with_wt_or_mut=True
        )
        
        # Tile start.
        tile_start = 0
        
        # Get tile length.
        tile_size = sequence_name_and_description.genomic_end_position - sequence_name_and_description.genomic_start_position + 1
        
        # Tile end.
        tile_end = tile_size
        
        sequence_record_without_adaptors = SeqRecord(
            seq=sequence_record_with_adapters.seq[len(prefix_seq):-len(suffix_seq)],
            id=seq_name,
            name=seq_name,
            description=seq_description
        )
        
        return Tile(sequence_record=sequence_record_without_adaptors,
                   tile_start=tile_start,
                   tile_end=tile_end,
                   prefix_seq=prefix_seq,
                   suffix_seq=suffix_seq
                  )
    
    def __init__(self, sequence_record, tile_start, tile_end, prefix_seq, suffix_seq):
        # Get parse info from sequence name and sequence description of original sequence record.
        sequence_name_and_description = SequenceNameAndDescription.from_str(
            seq_name=sequence_record.name,
            seq_description=sequence_record.description
        )
        

        # One-based genomic end coordinate of the tile sequence.
        sequence_name_and_description.genomic_end_position = sequence_name_and_description.genomic_start_position + tile_end
        # One-based genomic start coordinate of the tile sequence.
        sequence_name_and_description.genomic_start_position += tile_start
        
        # Filter out mutations that do not fall in the genomic interval of the tile.
        sequence_name_and_description.filter_mutations()
        print('before:', sequence_name_and_description.get_sequence_name_and_description(
            with_wt_or_mut=False
        ))
        # Get new sequence name and description based on the new genomic coordinates and filtering of mutations.
        self.seq_name, self.seq_description = sequence_name_and_description.get_sequence_name_and_description(
            with_wt_or_mut=True
        )
        print('after:', sequence_name_and_description.get_sequence_name_and_description(
            with_wt_or_mut=True
        ))
        print(self.seq_name)
        
        
        self.chrom = sequence_name_and_description.chrom
        self.genomic_start_position = sequence_name_and_description.genomic_start_position
        self.genomic_end_position = sequence_name_and_description.genomic_end_position
        
        
        # Create tile sequence.
        self.sequence = sequence_record.seq[tile_start:tile_end]
        
        self.sequence_record_with_adaptors = SeqRecord(
            seq=Seq(prefix_seq, generic_dna) + self.sequence + Seq(suffix_seq, generic_dna),
            id=self.seq_name,
            name=self.seq_name,
            description=self.seq_description
        )
        
        sequence_with_adaptors_md5 = hashlib.md5()
        sequence_with_adaptors_md5.update(str(self.sequence_record_with_adaptors.seq).encode('UTF-8'))
        self.md5 = sequence_with_adaptors_md5.hexdigest()

    def get_sequence_record(self):
        return SeqRecord(
            seq=self.sequence,
            id=self.seq_name,
            name=self.seq_name,
            description=self.seq_description
        )
    
    def get_sequence_record_with_adaptors(self):
        return self.sequence_record_with_adaptors
    
    def __repr__(self):
        return '<Tile>\n' \
               '  sequence: {0:s}\n' \
               '  seq_name: {1:s}\n' \
               '  seq_description: {2:s}\n' \
               '  chrom: {3:s}\n' \
               '  genomic_start_position: {4:d}\n' \
               '  genomic_end_position: {5:d}\n' \
               '  md5: {6:s}\n'.format(
                   str(self.sequence),
                   self.seq_name,
                   self.seq_description,
                   self.chrom,
                   self.genomic_start_position,
                   self.genomic_end_position,
                   self.md5
               )


sam_header = '''@HD	VN:1.0	SO:unsorted
@SQ	SN:chr1	LN:249250621
@SQ	SN:chr2	LN:243199373
@SQ	SN:chr3	LN:198022430
@SQ	SN:chr4	LN:191154276
@SQ	SN:chr5	LN:180915260
@SQ	SN:chr6	LN:171115067
@SQ	SN:chr7	LN:159138663
@SQ	SN:chr8	LN:146364022
@SQ	SN:chr9	LN:141213431
@SQ	SN:chr10	LN:135534747
@SQ	SN:chr11	LN:135006516
@SQ	SN:chr12	LN:133851895
@SQ	SN:chr13	LN:115169878
@SQ	SN:chr14	LN:107349540
@SQ	SN:chr15	LN:102531392
@SQ	SN:chr16	LN:90354753
@SQ	SN:chr17	LN:81195210
@SQ	SN:chr18	LN:78077248
@SQ	SN:chr19	LN:59128983
@SQ	SN:chr20	LN:63025520
@SQ	SN:chr21	LN:48129895
@SQ	SN:chr22	LN:51304566
@SQ	SN:chrX	LN:155270560
@SQ	SN:chrY	LN:59373566
@SQ	SN:chrM	LN:16571
@PG	ID:cheq_seq_tiler	PN:cheq_seq_tiles	VN:1.0.0'''


def write_tile_as_sam_record(sam_fh, tile):
    seq_name = tile.seq_name + ('___' + '___'.join(tile.seq_description.split(' '))
                                if tile.seq_description
                                else '')
    
    sequence = str(tile.sequence)
    
    # Make CIGAR string: Ignore mismatches for now.
    cigar = '{0:d}M'.format(len(sequence))
    
    # Make quality score for sequence (phred: 40)
    qual = 'I' * len(sequence)

    print(
        seq_name[0:250],             # QNAME: limit sequence name to 250 characters so the SAM file can be converted to BAM
        0,                           # FLAG: no mate and on the positive strand
        tile.chrom,                  # RNAME
        tile.genomic_start_position, # POS
        40,                          # MAPQ
        cigar,                       # CIGAR
        '*',                         # RNEXT: no mate
        0,                           # PNEXT: no mate
        0,                           # TLEN: no mate
        sequence,                    # SEQ
        qual,                        # QUAL
        sep='\t',
        end='\n',
        file=sam_fh
    )


def destroy_motif_matches_in_sequences(cbust_bed_filename,
                                       motif_filename,
                                       max_nbr_mutations,
                                       fasta_input_filename,
                                       fasta_output_filename):
    # Read Cluster-Buster BED output.
    cbust_bed = CbustBED.load_motifs_only_from_cbust_bed_from_file(cbust_bed_filename)
    
    # Read motif.
    motif = Motif.load_motif_from_file(
        motif_filename=motif_filename
    )

    # Read FASTA sequences.
    mut_fasta_sequences = MutateFastaSequences.load_fasta_from_file(
        fasta_input_filename=fasta_input_filename,
        motif_name=motif.motif_name
    )
    
    # Read each motif sequence match from Cluster-Buster output and break the motif.
    for cbust_motif_only in cbust_bed.cbust_motif_only_df[['seq_name',
                                                           'genomic_start__bed',
                                                           'relative_start__bed',
                                                           'strand',
                                                           'motif_sequence']].itertuples():
        motif.destroy_motif_in_fasta(
            mut_fasta_sequences=mut_fasta_sequences,
            seq_name=cbust_motif_only.seq_name,
            genomic_start=cbust_motif_only.genomic_start__bed,
            relative_start=cbust_motif_only.relative_start__bed,
            strand=cbust_motif_only.strand,
            motif_sequence=cbust_motif_only.motif_sequence,
            max_nbr_mutations=max_nbr_mutations
        )

    # Write mutated sequences to FASTA output file.
    mut_fasta_sequences.write_fasta_to_file(fasta_output_filename=fasta_output_filename)
```

```bash
import os

from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from Bio.Alphabet import generic_dna


os.chdir('/staging/leuven/stg_00002/lcb/dmauduit_HPC/CHEQseq_synth/Tile_generation')


shuffled_sequences_fasta_files_with_adapters_filenames = {
    'A': 'tileA.shuffled.with_adapters.unique.fa',
    'B': 'tileB.shuffled.with_adapters.unique.fa',
}

shuffled_sequences_fasta_files_with_adapters_no_motifs_filenames = {
    'A': 'tileA.shuffled.with_adapters.unique.no_motifs.fa',
    'B': 'tileB.shuffled.with_adapters.unique.no_motifs.fa',
}

shuffled_sequence_names_with_adapters_to_remove_filenames = {
    'A': 'tileA.shuffled.with_adapters.sequence_names_to_remove_because_they_have_motif.txt',
    'B': 'tileB.shuffled.with_adapters.sequence_names_to_remove_because_they_have_motif.txt',
}


for tile_kind in ('A', 'B'):
    nbr_sequences = 0
    seq_names_to_remove = set()
    
    with open(shuffled_sequence_names_with_adapters_to_remove_filenames[tile_kind], 'r') as shuffled_sequence_names_with_adapters_to_remove_fh:
        for line in shuffled_sequence_names_with_adapters_to_remove_fh:
            seq_names_to_remove.add(line.rstrip())

    # Write all shuffled sequences without motif.
    with open(shuffled_sequences_fasta_files_with_adapters_no_motifs_filenames[tile_kind], 'w') as shuffled_sequences_fasta_files_with_adapters_no_motifs_fh:
        with open(shuffled_sequences_fasta_files_with_adapters_filenames[tile_kind], 'r') as shuffled_sequences_fasta_files_with_adapters_fh:
            for sequence_record in SeqIO.parse(shuffled_sequences_fasta_files_with_adapters_fh, 'fasta'):
                if sequence_record.name not in seq_names_to_remove and nbr_sequences < 800:
#                     SeqIO.write(
#                         sequence_record,
#                         shuffled_sequences_fasta_files_with_adapters_no_motifs_fh,
#                         'fasta'
#                     )
                    print(
                        '>' + sequence_record.name,
                        str(sequence_record.seq),
                        sep='\n',
                        file=shuffled_sequences_fasta_files_with_adapters_no_motifs_fh
                    )
                    
                    nbr_sequences += 1
```
