## HOMER for SOX10 KD library

### Active vs inactive regions

__Format files for Homer__

```bash
# Convert the tiles to bed format
cd /staging/leuven/stg_00002/lcb/lcb_projects/CSE/SOX10-KD_library/analysis/HOMER/data

for file in *-human-names.txt; do \
 cat $file | tr "@@" "\t"  | tr ":" "\t" | tr "-" "\t" | awk '{print $1 "\t" $2 "\t" $3 "\t" $1 ":" $2 "-" $3 "@@" $4 }' > ${file%%.*}.bed ; done

# Remove entries that are mutated and the shuffled tiles
for file in *-human-names.bed; do \
 cat $file | grep -v "Neg-Control" > ${file%%.*}.noNC.bed ; done

# Convert bed to fasta
module load BEDTools
for file in *.noNC.bed; do \
bedtools getfasta \
-fi /staging/leuven/stg_00002/lcb/resources/human/hg19/Homo_sapiens_assembly19_sorted.fa \
-bed $file -name -fo ${file%%.*}.noNC.fa ; done

rm *.bed
```

__Run HOMER__

```bash
cd /staging/leuven/stg_00002/lcb/lcb_projects/CSE/SOX10-KD_library/analysis/HOMER/data
module load HOMER

findMotifs.pl \
active-human-names.noNC.fa \
fasta \
results/ \
-mask \
-nomotif \
-len 6,8,10,11,12,17,18 \
-fastaBg inactive-human-names.noNC.fa
 
```
